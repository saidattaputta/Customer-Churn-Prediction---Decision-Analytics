### What is customer churn?

Customer churn is when a customer stops doing business with a company, resulting in the loss of that customer. It is commonly measured as the percentage of customers lost over a given time period.

### Why is customer churn a BUSINESS problem and not just a data problem?

Customer churn is a business problem because losing customers directly affects revenue and growth, and keeping an existing customer is cheaper than acquiring a new one, while data only helps guide retention decisions.

### What exact decision will this project help the company make?

This project helps the company decide which existing customers should be contacted first with retention offers and support before they leave. It also guides where to focus the retention budget by identifying the customer groups that need immediate attention instead of spending equally on all customers.

### Who is the end user of this analysis or model?

The primary end user of this analysis is the **customer retention or CRM manager**, who is responsible for running retention campaigns. This role uses the results to decide which customers to prioritize for outreach and targeted offers each week.

### What is worse for the business in churn prediction?

It is worse for the business to **miss a customer who leaves without any intervention**, because that loss of future revenue is usually permanent. Offering a discount to a customer who would have stayed anyway only wastes some short-term cost, which is far less damaging than losing the customer entirely.
