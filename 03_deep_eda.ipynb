```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
raw_data_path = "data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv"
print("Sucessfully path loaded")
```

```python
df = pd.read_csv(raw_data_path)
pd.set_option('display.max_columns',None)
df.head(5)
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
df.info()
```

```python
df.head(5)
```

```python
sns.set(style = 'whitegrid')
```

## Churn vs Tenure

```python
plt.figure(figsize = (6,4))
sns.boxplot(x = 'Churn',y = 'tenure', data = df)
plt.title('Churn vs Tenure')
plt.show()
```

```python
plt.figure(figsize = (6,4))
sns.barplot(x = 'Churn',y = "tenure", data = df)
plt.title("Churn vs Avg Tenure")
plt.show()
```

### Data Inperpretation

- Customers with shorter tenure show significantly higher churn, indicating early-stage customers are more vulnerable to leaving.
- Customers with longer tenure are more likely to stay, suggesting increased commitment.
- This pattern implies that retention efforts should focus on customers in their initial months with the company.
- Customers in the very early tenure range show disproportionately high churn, though this may be influenced by small sample size and requires careful validation.


## Churn vs Contract

```python
plt.figure(figsize = (6,4))
sns.countplot(x = 'Contract', hue = 'Churn', data = df)
plt.title("Contract vs Churn")
plt.show()
```

### Data Interpretation

- Month-to-Month contracts show the highest churn, with a large number of customers leaving compared to other contract types.
- Customers on One-year contracts have significantly lower churn, indicating improved customer stability once a long-term commitment is made.
- Two-year contracts have the lowest churn, showing the strongest customer retention among all contract types.
- As contract duration increases, churn consistently decreases, making contract type a strong predictor of churn behavior.

## Churn vs Monthly Charges

```python
plt.figure(figsize = (6,4))
sns.boxplot(x= 'Churn', y= 'MonthlyCharges', data = df)
plt.title("Churn vs Monthly Charges")
plt.show()
```

### Data Interpretations

- Customers who churn have a higher median Monthly Charge compared to customers who stay.
- The upper range of Monthly Charges is higher for churned customers, indicating that high-paying customers are more likely to leave.
- Customers who do not churn show a wider spread toward lower Monthly Charges, suggesting price-sensitive customers tend to stay longer.
- There is a noticeable shift in distribution toward higher charges for churned customers.
